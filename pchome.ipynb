```python
import requests as req
import pandas as pd
import json as js
```

```python
url = 'https://ecshweb.pchome.com.tw/search/v3.3/all/results?q=%E4%BC%91%E9%96%92%E9%9B%B6%E9%A3%9F&page=1&sort=sale/dc'
headers = {"User-Agent":
           "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36"}
```

```python
res = req.get(url, headers = headers)
j = js.loads(res.text)
```

```python
j
```

```python
titles = []
prices = []
hrefs = []
fullpath = "https://24h.pchome.com.tw/prod/"

for pg in range(1,21):
    fullhref = "https://ecshweb.pchome.com.tw/search/v3.3/all/results?q=%E4%BC%91%E9%96%92%E9%9B%B6%E9%A3%9F&page=" + str(pg) + "&sort=sale/dc"
    r = req.get(fullhref, headers = headers)
    r_text = r.text
    if r_text == True:
        j = js.loads(r_text)
        datas = j["prods"]
    else:
        pass
    
    for i in datas:
        t = i["name"]
        p = i["price"]
        h = fullpath + i["Id"]

        titles.append(t)
        prices.append(p)
        hrefs.append(h)
```

```python
print(titles)
print(prices)
print(hrefs)
```

```python
len(titles),len(prices),len(hrefs)
```

```python
data = [titles, prices, hrefs]
dex = ['品名', '價格', '網址']
df = pd.DataFrame(data, index = dex, columns = list(range(1, len(titles)+1)))
```
